```python
%load_ext autoreload
%autoreload 2
import sys

sys.path.append("../")

from typing import List

import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
plt.rcParams["figure.figsize"] = [20, 10]

# import the ZOD DevKit
from zod import ZodFrames

# import default constants
import zod.constants as constants
from zod.constants import Camera, Lidar, Anonymization, AnnotationProject

# import useful data classes
from zod.data_classes import LidarData

# NOTE! Set the path to dataset and choose a version
dataset_root = "/content/drive/MyDrive/Colab Notebooks/zodNew"  # your local path to zod
version = "full"  # "mini" or "full"

# initialize ZodFrames
zod_frames = ZodFrames(dataset_root=dataset_root, version=version)
```

```python
# get default training and validation splits
training_frames = zod_frames.get_split(constants.TRAIN)
validation_frames = zod_frames.get_split(constants.VAL)

# print the number of training and validation frames
print(f"Number of training frames: {len(training_frames)}")
print(f"Number of validation frames: {len(validation_frames)}")

# print out the first 5 training frames
print("The 5 first training frames have the ids:", sorted(list(training_frames))[:5])
```

```python
zod_frame = zod_frames[6459]

# we can access the metadata of a frame
metadata = zod_frame.metadata

# print a subsample of meta data
print(f"Frame id: {metadata.frame_id}")
print(f"Country Code: {metadata.country_code}")
print(f"Time of day: {metadata.time_of_day}")
print(f"weather: {metadata.scraped_weather}")
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# Collect country codes for all frames
country_codes = []
all_frames = training_frames.union(validation_frames)

for frame_id in all_frames:
    frame = zod_frames[frame_id]
    country_codes.append(frame.metadata.country_code)

# Count occurrences of each country code
country_code_counts = Counter(country_codes)

# Prepare data for pie chart
labels = list(country_code_counts.keys())
sizes = list(country_code_counts.values())

# Plot pie chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Set title
plt.title('Geographic Diversity of the Dataset (Country Codes)')

# Display the plot
plt.show()
```

```python
import pandas as pd
from collections import Counter

# Collect country codes for all frames
country_codes = []
all_frames = training_frames.union(validation_frames)

for frame_id in all_frames:
    frame = zod_frames[frame_id]
    country_codes.append(frame.metadata.country_code)

# Count occurrences of each country code
country_code_counts = Counter(country_codes)

# Create a DataFrame from the counts
country_code_df = pd.DataFrame.from_dict(country_code_counts, orient='index', columns=['Frame Count'])
country_code_df.index.name = 'Country Code'

# Reset index to make 'Country Code' a column
country_code_df = country_code_df.reset_index()

# Display the DataFrame
print(country_code_df)

# Optionally, you can display the DataFrame in a more readable table format
from IPython.display import display
display(country_code_df)
```

```python
import pandas as pd
from collections import Counter

# Collect metadata details for all frames
metadata_list = []
all_frames = training_frames.union(validation_frames)

for frame_id in all_frames:
    frame = zod_frames[frame_id]
    metadata = frame.metadata
    metadata_list.append({
        'Country Code': metadata.country_code,
        'Weather Condition': metadata.time_of_day,
    })

# Create a DataFrame from the metadata
metadata_df = pd.DataFrame(metadata_list)

```
